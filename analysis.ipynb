### FIltering Data

#### Using loc[]

```python
#df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
new_df = df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Grass')]
new_df.head()
```

```python
# reset index (dropping the old index), inplace=True to save memory
new_df.reset_index(drop=True, inplace=True)
new_df.head()
```

#### Using str.contains()

```python
# return rows contains "Mega" in "Name" column
df.loc[df['Name'].str.contains('Mega')]

# ~: reverse the result
df.loc[~df['Name'].str.contains('Mega')]
```

#### Using str.contains() + regex

```python
import re

# flags=re.I: case insensitive, regex=True: using regex
#df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]

# ^pi: starting from "pi", [a-z]*: zero or more a-z letter, 
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

### Aggregate Statistics (Groupby)

```python
# return average defense of pokemons by types
df.groupby(['Type 1']).mean(numeric_only=True).sort_values('Sp. Atk', ascending=False)
```

```python
# add new column "count"
df['count'] = 1
df

# count by group "Type 1" & "Type 2"
df.groupby(['Type 1', 'Type 2']).count()['count']
```

### Working with large amounts of data

```python
# new_df: a new empty dataframe that contains the exact same columns as df
new_df = pd.DataFrame(columns=df.columns)

# for every five rows, apply perform .count() by each groups "Type 1"
# chunksize=5: 5 rows at a time
# the first 2 rows is the result from the first chunk (Four Grass Type 1 & one Fire Type 1)
for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()
    new_df = pd.concat([new_df, results])

new_df
```
